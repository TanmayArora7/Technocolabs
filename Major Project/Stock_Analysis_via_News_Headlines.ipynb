## Merging the dataframes(df1 and df2)

```python
# Merge the data
```

```python
merge = df1.merge(df2,how='inner', on='Date', left_index=True)
```

```python
merge.head()
```

## Data Cleaning for Headlines

```python
# Combine the top headlines

headlines = []

for row in range(0,len(merge.index)):
  headlines.append(' '.join(str(x) for x in merge.iloc[row,2:27]))
```

```python
#combined headline
headlines[0]
```

```python
# Data Cleaning
clean_headlines = []

for i in range(0,len(headlines)):
  clean_headlines.append(re.sub("b[(')]", '', headlines[i])) # remove b'
  clean_headlines[i] = re.sub('b[(")]', '', clean_headlines[i]) # remove b''
  clean_headlines[i] = re.sub("\'", '', clean_headlines[i]) # removes \'
```

```python
# combined clean headlines
clean_headlines[0]
```

```python
#Add clean headlines to the merge data set
merge['Combined_News'] = clean_headlines

merge['Combined_News'][0]
```

```python
# Show merge data 
merge.head()
```

## Create a function to get Subjectivity and Polarity and merging it to the dataframe

- Sentiment analysis models focus on **polarity** (positive, negative, neutral).
- Language can contain expressions that are objective or subjective. Objective expressions are facts. **Subjective** expressions are opinions that describe people’s feelings towards a specific subject or topic.

```python
# create a function to get the subjectivity
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity

# create a function to get the polarity
def getPolarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
#Create columns for Subjectivity and Polarity
merge['Subjectivity'] = merge['Combined_News'].apply(getSubjectivity)
merge['Polarity'] = merge['Combined_News'].apply(getPolarity)
```

```python
# show the columns in data 
merge.head()
```

## Creating a function to obtain sentiment score

```python
# Creating the sentiment scores
def getSIA(text):
  sia = SentimentIntensityAnalyzer()
  sentiment = sia.polarity_scores(text)
  return sentiment
```

```python
# Get the sentiment score for each day
compound = []
neg = []
pos = []
neu = []

SIA = 0

for i in range(0,len(merge['Combined_News'])):
  SIA = getSIA(merge['Combined_News'][i])
  compound.append(SIA['compound'])
  neg.append(SIA['neg'])
  neu.append(SIA['neu'])
  pos.append(SIA['pos'])
```

```python
#store the sentiment score in the merge data
merge['Compound'] = compound
merge['Negative'] = neg
merge['Neutral'] = neu
merge['Positive'] = pos
```

```python
#show the data
merge.head()
```

## Filtering out the columns which would be further used for prediction

```python
# Create a list of columns to be further used for prediction

keep_columns = ['Open', 'High', 'Low', 'Volume', 'Subjectivity', 'Polarity', 'Compound', 'Negative', 'Neutral', 'Positive', 'Label']
df = merge[keep_columns]
df.head()
```

## Finalizing feature and target data

```python
#Create the feature for data 
X = df
X = np.array(X.drop(['Label'],1))

#Create the target data
y = np.array(df['Label'])
```

## Linear Discriminant Analysis 

- Linear Discriminant Analysis, or LDA, uses the information from both(selection and target) features to create a new axis and projects the data on to the new axis in such a way as to minimizes the variance and maximizes the distance between the means of the two classes.
- Both LDA and PCA are linear transformation techniques: LDA is a supervised whereas PCA is unsupervised – PCA ignores class labels. LDA chooses axis to maximize the distance between points in different categories. 
- PCA performs better in case where number of samples per class is less. Whereas LDA works better with large dataset having multiple classes; class separability is an important factor while reducing dimensionality. 
- Linear Discriminant Analysis fails when the covariances of the X variables are a function of the value of Y. 

```python
#80-20 Split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#create and train the model
model = LinearDiscriminantAnalysis().fit(x_train,y_train)
```

```python
#show the predictions
predictions = model.predict(x_test)
predictions
```

```python
y_test
```

## Evaluate the model

**Accuracy , Precision , Recall , F1 Score and Specificity**
- Accuracy: What proportion of actual positives and negatives is correctly classified?
- Precision: What proportion of predicted positives are truly positive ?
-  Recall: What proportion of actual positives is correctly classified ?
-  F1 Score : Harmonic mean of Precision and Recall

```python
#show the model metrices
print(classification_report(y_test, predictions))
```
